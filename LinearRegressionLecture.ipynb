# Below should be read from csv files. Please change

```python
maindf = pd.read_csv('LinearRegressdata1.csv')
maindf.head()
```

```python
sns.pairplot(maindf)
```

```python
#Linear Model Starts Here
```

```python
from sklearn.model_selection import train_test_split
from sklearn import linear_model
```

```python
X_train, X_test, y_train, y_test = train_test_split(pd.DataFrame(maindf['X']), maindf['Y'], test_size=0.25)
print(type(X_train))
print(type(y_train))
```

```python
lmodelfirst = linear_model.LinearRegression()
lmodelfirst
```

```python
lmodelfirst.fit(X_train, y_train) #you can also do linear_model.LinearRegression().fit(X_train, y_train)
```

```python
ypredicted = lmodelfirst.predict(X_test)
```

```python
compareoutput = pd.DataFrame([ypredicted, y_test]).T
compareoutput.columns = ['Predicted', 'TestValue']
```

```python
compareoutput.head(10)
```

```python
import sklearn.metrics as metrics
```

```python
print('Coefficients of the Equation y = f(x) ', lmodelfirst.coef_)
print('Intercepts of the Equation y = f(x) ', lmodelfirst.intercept_)
print('Returns the Coefficient of Determination ' , lmodelfirst.score(X_train, y_train))
print('Model Parameters ' , lmodelfirst.get_params(deep=True))
mse = np.mean((ypredicted-y_test)**2)
```

```python
print('explained_variance: ', metrics.explained_variance_score(y_test, ypredicted))    
print('r2: ', metrics.r2_score(y_test, ypredicted))
print('MAE: ', round(metrics.mean_absolute_error(y_test, ypredicted),2))
print('MSE: ', round(metrics.mean_squared_error(y_test, ypredicted), 2))
print('RMSE: ', round(np.sqrt(mse),4))
```

# Second Linear Regression Model

```python
maindfywithnoise = pd.read_csv('LinearRegressdata2.csv')
maindfywithnoise.head()
```

```python
sns.pairplot(maindfywithnoise)
```

```python
X_train, X_test, y_train, y_test = train_test_split(pd.DataFrame(maindfywithnoise['X']), maindfywithnoise['Y'], test_size=0.25)
lmodelsecond = linear_model.LinearRegression()

ypredicted = lmodelsecond.fit(X_train, y_train).predict(X_test) 
yerror = ypredicted-y_test
print('Coefficients of the Equation y = f(x) ', lmodelsecond.coef_)
print('Intercepts of the Equation y = f(x) ', lmodelsecond.intercept_)
print('Returns the Coefficient of Determination ' , lmodelsecond.score(X_train, y_train))
print('Model Parameters ' , lmodelsecond.get_params(deep=True))
mse = np.mean((yerror)**2)
rmse = np.sqrt(mse)
print(rmse)
```

```python
compareoutput = pd.DataFrame([X_test['X'].values, ypredicted, y_test.values, yerror.values]).T
compareoutput.columns = ['X', 'Predicted', 'Y', 'Error']
compareoutput['SrNo'] = compareoutput.index
compareoutput.head(3)
```
